```python
data=pd.read_csv("Salary_Data.csv")
data
```

```python
x = data['YearsExperience']
y = data['Salary']
```

```python
data.corr()
```

```python
b, m = polyfit(x, y, 1)
plt.scatter(x, y)
plt.plot(x, y, '.')
plt.plot(x, b + m * x, '-')
plt.title('Scatter plot Salary Hike')
plt.xlabel('Years of Experience')
plt.ylabel('Salary Hike')
plt.show()
```

```python
model = sm.OLS(y, x).fit()
predictions = model.predict(x)
model.summary()
```
